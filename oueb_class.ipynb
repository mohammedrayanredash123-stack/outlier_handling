```python
port_stem = PorterStemmer()
stop_words = set(stopwords.words('english'))

def stemming(content):
    # Remove non-alphabet and lowercase
    content = re.sub('[^a-zA-Z]', ' ', content).lower().split()

    # Remove stopwords and stem
    content = [port_stem.stem(word) for word in content if word not in stop_words]

    return ' '.join(content)
```

```python
from tqdm import tqdm #tqdm = "taqaddum" (Arabic word meaning "progress")
tqdm.pandas()

# Assuming stemming function is defined in a previous cell
news_dataset['content'] = news_dataset['content'].progress_apply(stemming)
#cleans entire dataset.
```
